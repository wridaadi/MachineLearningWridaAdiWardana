## Introduction
Machine Learning merupakan ilmu cabang dari kecerdasan buatan yang mempelajari bagaimana cara membuat komputer agar belajar dari data yang diberikan sehingga dapat mengenali suatu objek atau memprediksi sesuatu berdasarkan data yang ada.


## Proses Machine Learning
- __Data__, Data yang digunakan berupa data input dan target yang nanti akan digunakan dalam proses pembelajaran pada mesin.
- __Algoritma Learning__, Algoritma learning digunakan sebagai metode dalam proses pembelajaran.
- __Pelatihan__, Proses pelatihan dilakukan dengan mmenggunkan data dan menerapkan algoritma learning pada program.
- __Model__, Setelah dilakukan proses pelatihan maka akan didapatkan pemodelan yang akan digunakan untuk pengenalan, prediksi, dan lain sebagainya.

## Penggunaan Machine Learning
Penggunaan machine learning dalam kehidupan sehari-hari adalah sebagai berikut:
- Recognition
- Detection
- Recomendation
- Prediction

## Metode Machine Learning
Cara yang digunakan dalam Machine Learning terbagi menjadi dua, yaitu:
- __Supervised Learning__, Proses Supervised Learning melibatkan pengawas yang berfungsi untuk mengontrol output dari program dengan menggunakan data yang dapat dijadikan target output.
- __Unsupervised Learning__, Proses Unsupervised Learning tidak melibatkan pengawas sehingga membiarkan komputer belajar sendiri, cara ini digunakan untuk memfilter atau mengelompokkan data berdasarkan kemiripan fitur.

##### Gradient Descent
*Gradient Descent* digunakan untuk mencari nilai koefisien dari model Multivariate Regression dengan cara meminimalkan error dari model yang dihasilkan oleh Data Training. Berikut tahapan pada Gradient Descent:
- Menghitung error yang dihasilkan oleh data set (antara output dengan target)
- Tentukan seberapa jauh perubahan menuju error minimal dengan menggunakan *Learning rate*
- Lakukan secara berulang sampai error minimal tercapai
